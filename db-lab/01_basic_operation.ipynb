### Column filtering

```python
df_column_filtered = df.select("title","imdb_rating","industry")
# Truncate when set to False displays the entire value of the column
df_column_filtered.show(3, truncate=False)
```

### Row Filtering

```python
df_filtered = df.filter((df["release_year"] >= 2000) & (df["release_year"] <= 2010))
display(df_filtered)
```

```python
from pyspark.sql.functions import col
df_filtered = df.filter((col("release_year") >= 2000) & (col("release_year") <= 2010))
display(df_filtered)
```

```python
df.filtered = df.filter(col("release_year").between(2000, 2010))
display(df.filtered)
```

```python
unique_industry_df = df.select("language").distinct()
display(unique_industry_df)
```

```python
df = df.withColumn('profit', col('revenue') - col("budget"))
display(df)
```

```python
df = df.withColumnRenamed('revenue','total_revenue')
df.printSchema()
```
