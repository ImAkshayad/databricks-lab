```python
# Manually define the schema 
from pyspark.sql import types as T 

csv_schema = T.StructType([
    T.StructField("order_date", T.TimestampType(), True),
    T.StructField("country", T.StringType(), True),
    T.StructField("order_id", T.IntegerType(), True),
    T.StructField("product", T.StringType(), True),
    T.StructField("qty", T.IntegerType(), True),
    T.StructField("price", T.DoubleType(), True)
])

df = spark.read \
        .option("header", True) \
            .option("dateFormat", "yyyy-dd-MM") \
                .schema(csv_schema) \
                    .csv('/Volumes/workspace/default/raw_layer/orders.csv')
display(df)
df.printSchema()
```

```python
# Write functionality
df.write.mode("overwrite").parquet("/Volumes/workspace/default/raw_layer/orders")
```
